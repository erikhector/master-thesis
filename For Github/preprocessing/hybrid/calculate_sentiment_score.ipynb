```python
# filepath = "../../data/twitter/alphabet_tweets_classified_with_label.csv"
# filepath = "../../data/twitter/microsoft_tweets_classified_with_label.csv"
filepath = "../../data/twitter/apple_tweets_classified_with_label.csv"
```

```python
import pandas as pd


df = pd.read_csv(filepath,lineterminator='\n', usecols=['created_at', 'text', 'predictions', 'label'])
```

```python
df.sample(100)
```

```python
grouped_df = df.groupby(['created_at','label']).size().reset_index(name='counts')
```

```python
grouped_df.sample(100)
```

```python
cleaned_df = pd.DataFrame(columns=['date', 'neutral', 'positive', 'negative'])
```

```python
def add_to_cleaned_df(val):
    date = val['created_at']
    count = val['counts']
    if (val['label'] == 'positive'):
        val['date'] = date
        val['positive'] = count      
    if (val['label'] == 'neutral'):
        val['date'] = date
        val['neutral'] = count 
    if (val['label'] == 'negative'):
        val['date'] = date
        val['negative'] = count 
    return val

final_df = grouped_df.apply(add_to_cleaned_df, axis=1)
```

```python
final_df.head()
```

```python
df_final = final_df.drop(columns=['created_at', 'label', 'counts'], inplace=True)
```

```python


final_df = final_df.fillna(0)

final_df.sample(100)
```

```python
grouped_final_df = final_df.groupby(by=['date']).sum()
```
